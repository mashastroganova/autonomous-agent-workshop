```python
import azure.cosmos.documents as documents
import azure.cosmos.cosmos_client as cosmos_client
import azure.cosmos.exceptions as exceptions
from azure.cosmos.partition_key import PartitionKey
import datetime
import random
import string

HOST = settings['host']
MASTER_KEY = settings['master_key']
DATABASE_ID = settings['database_id']
CONTAINER_ID = settings['container_id']

# Initialize the Cosmos client
client = cosmos_client.CosmosClient(HOST, {'masterKey': MASTER_KEY})

# Get the database and container
db = client.get_database_client(DATABASE_ID)
container = db.get_container_client(CONTAINER_ID)

# Function to generate random customer data
def generate_customer_data(customer_id):
    first_names = ['John', 'Emma', 'Michael', 'Sophia', 'William', 'Olivia', 'James', 'Ava', 'Benjamin', 'Isabella']
    last_names = ['Smith', 'Johnson', 'Brown', 'Taylor', 'Miller', 'Davis', 'Garcia', 'Wilson', 'Martinez', 'Anderson']
    service_tiers = ['Basic', 'Standard', 'Premium', 'Gold']
    internet_connection_types = ['DSL', 'Cable', 'Fiber', 'Satellite']
    connection_statuses = ['Active', 'Inactive', 'Pending', 'Cancelled']
    contract_terms = [12, 24, 36]
    phone_models = ['iPhone 12', 'Samsung Galaxy S21', 'Google Pixel 5', 'OnePlus 9 Pro', 'Xiaomi Mi 11']
    complaint_history = ['Last 3 days', 'Last week', 'Last month', 'Last 3 months', 'Last 6 months', 'Last year', 'More than a year ago', 'Never']
    
    full_name = f"{random.choice(first_names)} {random.choice(last_names)}"
    age = random.randint(18, 65)
    address = f"{random.randint(1, 999)} {random.choice(string.ascii_uppercase)} Street, {random.choice(['New York', 'Chicago', 'San Francisco', 'Munich', 'Berlin'])}"
    service_tier = random.choice(service_tiers)
    internet_connection_type = random.choice(internet_connection_types)
    contract_term = random.choice(contract_terms)
    monthly_bill = round(random.uniform(50, 200), 2)
    phone_model = random.choice(phone_models)
    connection_status = random.choice(connection_statuses)
    phone_number = f"+1{random.randint(100, 999)}{random.randint(100, 999)}{random.randint(1000, 9999)}"
    email = f"{full_name.replace(' ', '.').lower()}@{random.choice(['email1.com', 'email2.com', 'email3.com'])}"
    complaint_history = random.choice(complaint_history)
    
    return {
        'id': customer_id,
        'FullName': full_name,
        'Age': age,
        'Address': address,
        'ServiceTier': service_tier,
        'InternetConnectionType': internet_connection_type,
        'ContractTerm': contract_term,
        'MonthlyBill': monthly_bill,
        'PhoneModel': phone_model,
        'PhoneNumber': phone_number,
        'Email': email,
        'ConnectionStatus': connection_status,
        'ComplaintHistory': complaint_history,
    }

# Function to create and insert customer items
def create_customer_items(container, num_items):
    print(f"\nCreating {num_items} customer items\n")
    
    for i in range(1, num_items + 1):
        customer_id = f"Customer{i}"
        customer_data = generate_customer_data(customer_id)
        
        try:
            container.create_item(body=customer_data)
            print(f"Created customer item: {customer_id}")
        except exceptions.CosmosHttpResponseError as e:
            print(f"Error creating customer item {customer_id}: {e}")

# Create and insert mock customer data
num_customers = 30
create_customer_items(container, num_customers)
```
